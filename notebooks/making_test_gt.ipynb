```python
import os
import glob
import shutil
from scipy.misc import imread
```

```python
test_filenames = glob.glob('../../Datasets/Huawei/DATASET_INPAINTING/test/*')
train_filenames = glob.glob('../../Datasets/Huawei/DATASET_INPAINTING/train_gt/*')
```

```python
train_filenames_with_masks = glob.glob('../../Datasets/Huawei/DATASET_INPAINTING/train/*')
train_filenames_with_masks
```

```python
test_filenames = [f.split('/')[-1].split('.')[0] for f in test_filenames]
```

```python
for file in train_filenames:
    # Find corresponding ground truth files in train
    if file.split('/')[-1].split('_')[0] in test_filenames:
        # Copy file
        shutil.copy(file, '../../Datasets/Huawei/DATASET_INPAINTING/test_gt/')
```

```python
# Remove test files from test and test_gt
for file in train_filenames:
    # Find corresponding ground truth files in train
    if file.split('/')[-1].split('_')[0] in test_filenames:
        # Copy file
        print('Removing: ', file)
        os.remove(file)
```

```python
# Remove test files from test and test_gt
for file in train_filenames_with_masks:
    # Find corresponding ground truth files in train
    if file.split('/')[-1].split('.')[0] in test_filenames:
        # Copy file
        print('Removing: ', file)
        os.remove(file)
```
